## Opening the files, removing stop words, removing symbols, removing emojis and somewhat images

```python

texts = []
punctuation = ['(', ')', '?', ':', ';', ',', '.', '!', '/', '"', "'"] 
stop_words = nltk.corpus.stopwords.words('english')
symbols = set('  #  * ; , / \ ! | #/%'.split())
for p in file:
    with codecs.open(p, 'r', encoding='utf-8', errors='ignore') as fdata:
        contents = fdata.read().lower()
        
        L = contents.split()
        texts.append(L)        
```
